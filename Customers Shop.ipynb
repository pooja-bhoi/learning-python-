```python
sns.pairplot(df.drop("CustomerID", axis =1))
```

```python
sns.countplot(x ='Gender', data = df)
plt.title("Gender Data Distribution")
plt.show()
```

```python
sns.histplot(x = 'Age', data = df, color = 'purple', bins = 10)
plt.title("Customer Age Distribution")
plt.show()
```

```python
sns.kdeplot(x = 'Annual Income ($)', data=df, color='blue', fill=True)
plt.title("Income Data Distribution")
plt.show()
```

```python
sns.countplot(x ='Profession', data=df)
plt.xticks(rotation = 45)
plt.title("Profession Data Distribution")
plt.show()
```

```python
sns.kdeplot(x = 'Work Experience', data=df, color='black', fill =True)
plt.title("Work Experience Data Distribution")
plt.show()
```

```python
sns.countplot(x ='Family Size', data=df)
plt.title("Family Size Distribution")
plt.show()
```

```python
sns.violinplot(x= 'Gender', y = 'Spending Score (1-100)', data=df)
plt.title('Spending Score Gender Distribution')
plt.show()
```

```python
sns.lineplot(x = 'Age', y = 'Spending Score (1-100)', data =df, color='pink')
plt.title("Spending Age Distribution")
plt.show()
```

```python
sns.lineplot(x = 'Age', y = 'Spending Score (1-100)', hue ='Gender', data =df)
plt.title("Spending Score by Age and Gender")
plt.show()
```

```python
plt.hexbin(x = 'Annual Income ($)', y = 'Spending Score (1-100)', data=df, gridsize= 20, cmap ='Blues')
plt.xlabel('Annual Income ($)')
plt.xticks(rotation = 45)
plt.ylabel('Spending Score (1-100)')
plt.title('Spending Score by Income')
plt.colorbar()
plt.show()
```

```python
sns.barplot(x ='Profession', y ='Spending Score (1-100)', data=df)
plt.xticks(rotation = 45)
plt.title('Profession and Spending Score')
plt.show()
```

```python
sns.boxplot(x ='Work Experience', y = 'Spending Score (1-100)', data=df)
plt.title('Work Experience and Spending Score')
plt.show()
```

```python
sns.violinplot(x ='Family Size', y = 'Spending Score (1-100)', data=df)
plt.title ('Spending Score by Family Size')
plt.show()
```
